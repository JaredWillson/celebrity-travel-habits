```python

#import dependencies
from splinter import Browser
from bs4 import BeautifulSoup
import re
import csv
import time
from time import sleep
import json
import requests
```

```python

#set up the browser
browser = Browser('chrome')
url = 'https://celebrityprivatejettracker.com/leaderboard/'

#create the session
browser.visit(url)
```

```python
#intiate html session
html = browser.html
soup = BeautifulSoup(html, 'html.parser')
```

```python
#scraping the table (class is table4)

table_info = soup.find_all(class_='table4')
table_info

plane_links = soup.find_all('a', string=re.compile('^N'))

#create the empty list to be filled with the plane information
plane_list =[]

for plane in table_info:
    for link in plane.find_all('a', string=re.compile('^N')):
        plane_list.append(link)

# Extract URLs from the plane_list
url_list = [tag['href'] for tag in plane_list if tag and 'href' in tag.attrs]

print(url_list)
```

```python
#strip the plane registration from the link info
plane_registration = []
plane_reg_list = soup.find_all('a', string=re.compile('^N'))
for plane in plane_list:
    if plane not in ["Nike", "Nike Corporation"]:
        plane_registration.append(plane.get_text().strip())
    
plane_registration
```

```python
url_list = url_list

# Function to visit each URL
def visit_urls(url_list):
    for url in url_list:
        try:
            response = requests.get(url)
            # Check if the request was successful
            if response.status_code == 200:
                print(f"Successfully visited: {url}")
            else:
                print(f"Failed to visit: {url} with status code {response.status_code}")
        except requests.exceptions.RequestException as e:
            print(f"Error visiting {url}: {e}")
            
            
url_list
```

```python
#merge the url list with the plane registration list
merged_list = list(zip(url_list, plane_registration))

merged_list
```

```python
# Write the flight information to a CSV file

# Define the CSV file path
plane_information = 'plane_information.csv'

# Open the CSV file in append mode to add new entries
with open(plane_information, 'a', newline='') as myfile:
    wr = csv.writer(myfile, quoting=csv.QUOTE_ALL)

    # Check if the file is empty, if so, write the header row
    if myfile.tell() == 0:
        wr.writerow(['URL', 'ID'])

    # Write all entries from flight_info to the CSV file
    for flight in merged_list:
        wr.writerow(flight)
```

```python
#start the loop of scraping each link/page on the website

flight_info = []

for link, plane_id in merged_list:

    try:
        # Visit the link
        browser.visit(link)
        time.sleep(2)  # Allow time for the page to load

        # Try to find the button using different methods (i forget which one worked so I'm leaving them all in)
        button = browser.find_by_name('Sort by Flight Time')
        if not button:
            button = browser.find_by_text('Sort by Flight Time')
        if not button:
            button = browser.find_by_css('button[name="Sort by Flight Time"]')
        
        # Click the button if found
        if button:
            button.click()
            time.sleep(15)  # Allow time for the new page to load--it's slow so setting to 15s
        else:
            print("Button not found.")

        #begin fetching the content
        print("Fetching page content.")
        html = browser.html
        soup = BeautifulSoup(html, 'html.parser')


        # Extract the necessary information
        table = soup.find(class_='table3')
        if not table:
            continue
        rows = table.find_all('tr')
        #set up the variables
        flight_date = None
        airports = []
        flight_dist = None
        flight_time = None
        flight_fuel = None
        for tr in rows:
            #get the cells
            tds = tr.find_all('td')
            for td in tds:
                #make a text object for ease
                text = td.get_text(strip=True)
               
                #fetch flight date
                if not flight_date and re.match(r'\d{2}/\d{2}/\d{4}', text):
                    flight_date = text  
               
                # fetch airports
                a_elements = td.find_all('a', href=re.compile(r'/airport-iata'))
                for a_element in a_elements:
                    href = a_element.get('href')
                    match = re.search(r'airport-iata-(.*?)/', href)
                    if match:
                       extracted_value = match.group(1)
                       airports.append(extracted_value) 
                
                # Fetch flight distance
                if not flight_dist and 'miles' in text:
                    match = re.search(r'\b[\w\s,]+\bmiles\b', text)
                    if match:
                        flight_dist = match.group()

                #Fetch flight time
                if not flight_time and ('hours' in text or 'minutes' in text):
                    match = re.search(r'\b[\w\s,]+(?:hours|minutes)\b', text)
                    if match:
                        flight_time = match.group()

                #Fetch flight fuel usage
                if not flight_fuel and 'gallons' in text:
                    match = re.search(r'\b[\w\s,]+\bgallons\b', text)
                    if match:
                        flight_fuel = match.group()
        
        
            # Append the extracted information to the flight_info list
            flight_info.append([
                plane_id if plane_id else None,
                flight_date if flight_date else None,
                airports[0] if len(airports) > 0 else None,
                airports[1] if len(airports) > 1 else None,
                flight_dist if flight_dist else None,
                flight_time if flight_time else None,
                flight_fuel if flight_fuel else None
            ])
            flight_date = None
            airports = []
            flight_dist = None
            flight_time = None
            flight_fuel = None
            # print to make sure it's working as it should
            # print(f"Flight date: {flight_date if flight_date else 'N/A'}")
            # print(f"Departure airport: {airports[0] if len(airports) > 0 else 'N/A'}")
            # print(f"Arrival airport: {airports[1] if len(airports) > 1 else 'N/A'}")
            # print(f"Flight distance: {flight_dist if flight_dist else 'N/A'}")
            # print(f"Flight time: {flight_time if flight_time else 'N/A'}")
            # print(f"Flight fuel: {flight_fuel if flight_fuel else 'N/A'}")
        
        
    except Exception as e:
        print(f"An error occurred for link {url}: {e}")



print("Process completed.")
print("writing to CSV")
```

```python
# Write the flight information to a CSV file

# Define the CSV file path
flight_information = 'flight_info.csv'

with open('flight_info.csv', mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['Plane ID', 'Flight Date', 'Departure Airport', 'Arrival Airport', 'Flight Distance', 'Flight Time', 'Fuel Used'])
    writer.writerows(flight_info)
```
